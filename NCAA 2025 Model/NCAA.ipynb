```python
def GamePred(team1list,team2list, SinglePred = True):
    predict = {
        "Team1" : team1list,
        "Team2" : team2list
    }
    df_predict = pd.DataFrame(predict)
    firstmerge = pd.merge(df_predict,Predictdata, left_on="Team1", right_on='School', how='left')
    firstmerge = firstmerge.drop('School', axis='columns')
    secondmerge = pd.merge(firstmerge,Predictconf, left_on="Team1", right_on='School', how='left')
    secondmerge = secondmerge.drop('School', axis='columns')
    thridmerge = pd.merge(secondmerge,Predictdata, left_on="Team2", right_on='School', how='left')
    thridmerge = thridmerge.drop('School', axis='columns')
    fourthmerge = pd.merge(thridmerge,Predictconf, left_on="Team2", right_on='School', how='left')
    fourthmerge = fourthmerge.drop('School', axis='columns')
    #Ordianl
    enc = OrdinalEncoder()
    fourthmerge["Conf_x"] = enc.fit_transform(fourthmerge[["Conf_x"]])
    fourthmerge["Conf_y"] = enc.fit_transform(fourthmerge[["Conf_y"]])
    fourthmerge = fourthmerge.drop(columns="Team1")
    fourthmerge = fourthmerge.drop(columns="Team2")
    firstg_pred = model1.predict(fourthmerge)
    predict = {
        "Team1" : team2list,
        "Team2" : team1list
    }
    df_predict = pd.DataFrame(predict)
    firstmerge = pd.merge(df_predict,Predictdata, left_on="Team1", right_on='School', how='left')
    firstmerge = firstmerge.drop('School', axis='columns')
    secondmerge = pd.merge(firstmerge,Predictconf, left_on="Team1", right_on='School', how='left')
    secondmerge = secondmerge.drop('School', axis='columns')
    thridmerge = pd.merge(secondmerge,Predictdata, left_on="Team2", right_on='School', how='left')
    thridmerge = thridmerge.drop('School', axis='columns')
    fourthmerge = pd.merge(thridmerge,Predictconf, left_on="Team2", right_on='School', how='left')
    fourthmerge = fourthmerge.drop('School', axis='columns')
    #Ordianl
    enc = OrdinalEncoder()
    fourthmerge["Conf_x"] = enc.fit_transform(fourthmerge[["Conf_x"]])
    fourthmerge["Conf_y"] = enc.fit_transform(fourthmerge[["Conf_y"]])
    fourthmerge = fourthmerge.drop(columns="Team1")
    fourthmerge = fourthmerge.drop(columns="Team2")
    secondg_pred = model1.predict(fourthmerge)
    finalpredictlist = []
    for x in range(len(firstg_pred)):
        first = firstg_pred[x]
        second = secondg_pred[x]
        team1dif = (first[0] + second[1])/2
        team2dif = (first[1] + second[0])/2
        finalpredictlist.append([team1dif,team2dif])
    gamenum = 0
    if SinglePred == True:
        for x in finalpredictlist:
            print(f"Team1: {team1list[gamenum]} Team2: {team2list[gamenum]}")
            print(f"{x[0]} , {x[1]}")
            print("--------------------------")
            gamenum = gamenum + 1
    else:
        gamelist = []
        winnerlist = []
        for x in finalpredictlist:
            if x[0] > x[1]:
                winnerlist.append(team1list[gamenum])
            else:
                winnerlist.append(team2list[gamenum])
            gamelist.append(f"Team1: {team1list[gamenum]} {x[0]} Team2: {team2list[gamenum]} {x[1]}")
            gamenum = gamenum + 1
        return winnerlist, gamelist
```

```python
def Tour(EastT1, EastT2, MidWestT1, MidWestT2, SouthT1, SouthT2, WestT1, WestT2, FinalsMatchups):
    scoredict = {"East" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []},"MidWest" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []}
                 ,"South" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []},"West" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []}
                 ,"Finals" : {"Semi" : [], "Final" : []}}
    regionlist = [[EastT1, EastT2],[MidWestT1, MidWestT2],[SouthT1, SouthT2],[WestT1, WestT2]]
    regions = ["East", "MidWest", "South", "West"]
    rn = 0
    regionwinners = []
    for x in regionlist:
        winners, scores = GamePred(x[0],x[1],SinglePred=False)
        scoredict[regions[rn]]["Round1"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = GamePred(nrlt1,nrlt2,SinglePred=False)
        scoredict[regions[rn]]["Round2"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = GamePred(nrlt1,nrlt2,SinglePred=False)
        scoredict[regions[rn]]["Round3"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = GamePred(nrlt1,nrlt2,SinglePred=False)
        scoredict[regions[rn]]["Round4"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        regionwinners.append([nrlt1, regions[rn]])
        rn = rn + 1
    nrlt1 = []
    nrlt2 = []
    for x in FinalsMatchups:
        for y in regionwinners:
            if y[1] == x[0]:
                nrlt1.append(y[0][0])
            if y[1] == x[1]:
                nrlt2.append(y[0][0])
    winners, scores = GamePred(nrlt1,nrlt2,SinglePred=False)
    scoredict["Finals"]["Semi"] = scores
    nrlt1 = []
    nrlt2 = []
    for y in winners:
        if (nn % 2) == 0:
            nrlt1.append(y)
        else:
            nrlt2.append(y)
        nn = nn + 1
    winners, scores = GamePred(nrlt1,nrlt2,SinglePred=False)
    scoredict["Finals"]["Final"] = scores    
    return scoredict
```

```python
def randPec(x):
    my_array = np.array([-0.09,-0.05,-0.03,0.0,0.03,0.05,0.09])
    random_element = np.random.choice(my_array)
    x = x + random_element
    return x
```

```python
def sim(EastT1, EastT2, MidWestT1, MidWestT2, SouthT1, SouthT2, WestT1, WestT2, FinalsMatchups,num_of_sim = 1):
    scoredict = {"East" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []},"MidWest" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []}
                 ,"South" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []},"West" : {"Round1" : [], "Round2" : [], "Round3" : [], "Round4" : []}
                 ,"Finals" : {"Semi" : [], "Final" : []}}
    regionlist = [[EastT1, EastT2],[MidWestT1, MidWestT2],[SouthT1, SouthT2],[WestT1, WestT2]]
    regions = ["East", "MidWest", "South", "West"]
    rn = 0
    regionwinners = []
    for x in regionlist:
        winners, scores = NewGamePred(x[0],x[1],SinglePred=False,num_of_sim = num_of_sim)
        scoredict[regions[rn]]["Round1"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = NewGamePred(nrlt1,nrlt2,SinglePred=False,num_of_sim =num_of_sim)
        scoredict[regions[rn]]["Round2"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = NewGamePred(nrlt1,nrlt2,SinglePred=False,num_of_sim =num_of_sim)
        scoredict[regions[rn]]["Round3"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        winners, scores = NewGamePred(nrlt1,nrlt2,SinglePred=False,num_of_sim =num_of_sim)
        scoredict[regions[rn]]["Round4"] = scores
        nn = 0
        nrlt1 = []
        nrlt2 = []
        for y in winners:
            if (nn % 2) == 0:
                nrlt1.append(y)
            else:
                nrlt2.append(y)
            nn = nn + 1
        regionwinners.append([nrlt1, regions[rn]])
        rn = rn + 1
    nrlt1 = []
    nrlt2 = []
    for x in FinalsMatchups:
        for y in regionwinners:
            if y[1] == x[0]:
                nrlt1.append(y[0][0])
            if y[1] == x[1]:
                nrlt2.append(y[0][0])
    winners, scores = NewGamePred(nrlt1,nrlt2,SinglePred=False,num_of_sim =num_of_sim)
    scoredict["Finals"]["Semi"] = scores
    nrlt1 = []
    nrlt2 = []
    for y in winners:
        if (nn % 2) == 0:
            nrlt1.append(y)
        else:
            nrlt2.append(y)
        nn = nn + 1
    winners, scores = NewGamePred(nrlt1,nrlt2,SinglePred=False,num_of_sim =num_of_sim)
    scoredict["Finals"]["Final"] = scores    
    return scoredict
```
