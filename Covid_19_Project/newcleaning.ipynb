```python
def compgraph(df, county, wave):
    for i in graphlist:
        for u in graphlist:
            if i == u:
                pass
            else:
                try:
                    plt.clf()
                    coefficients = np.polyfit(df[i], df[u], 1)
                    m = coefficients[0]
                    b = coefficients[1]
                    y_reg = m * df[i] + b
                    plt.scatter(df[i],df[u])
                    plt.plot(df[i], y_reg, color='b', label='Regression Line')
                    plt.xlabel(i)
                    plt.ylabel(u)
                    plt.title("Scatter Plot with Regression Line")
                    plt.legend()
                    plt.savefig(f"county/{county}/{i}/{wave}+{i}+{u}.png")
                except:
                    pass
def timegraph(df, county, wave):
    for u in graphlist:
        plt.clf()
        plt.plot(df["As of Date"],df[u])
        plt.xlabel("Time")
        plt.ylabel(u)
        plt.title(f"{u} graphed for {wave}")
        plt.savefig(f"county/{county}/TimeSeries/{wave}+{u}.png")
        
```

```python
wavelist = ["Wave 1","Wave 2","Wave 3"]
```

```python
columnlist = []
for x in graphlist:
    for y in wavelist:
        columnlist.append(f"{y}_{x}")
```

```python
avedf = pd.DataFrame(columns=columnlist, index=countynames)
```

```python
countynames
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# 14 March 2020–28 June 2020
# 10 October 2020–1 June 2021
# 26 October 2021–8 March 2022
# https://coronavirus.jhu.edu/region/us/new-york
```

```python
for u in countynames:
    countydf = df[df["Facility County"] == u]
    countydf["As of Date"] = pd.to_datetime(countydf["As of Date"])
    countydf = countydf.drop(columns=["Facility PFI","Facility Name",	"DOH Region",	"Facility County",	"Facility Network",	"NY Forward Region"])
    countydf = countydf.groupby('As of Date')[graphlist].sum()
    countydf = countydf.reset_index()
    wave1df = countydf[(countydf["As of Date"] >= '2020-03-14') & (countydf["As of Date"] <= '2020-06-28')].copy()
    wave2df = countydf[(countydf["As of Date"] >= '2020-10-10') & (countydf["As of Date"] <= '2021-06-01')].copy()
    wave3df = countydf[(countydf["As of Date"] >= '2021-10-26') & (countydf["As of Date"] <= '2022-03-08')].copy()
    compgraph(wave1df,u, wavelist[0])
    compgraph(wave2df,u, wavelist[1])
    compgraph(wave3df,u, wavelist[2])
    timegraph(wave1df,u, wavelist[0])
    timegraph(wave2df,u, wavelist[1])
    timegraph(wave3df,u, wavelist[2])
    for x in graphlist:
        avedf.loc[u, f"{wavelist[0]}_{x}"] = wave1df[x].mean()
    for x in graphlist:
        avedf.loc[u, f"{wavelist[1]}_{x}"] = wave2df[x].mean()
    for x in graphlist:
        avedf.loc[u, f"{wavelist[2]}_{x}"] = wave3df[x].mean()
    print(f"{u} is done")
```
