```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from datetime import datetime
df = pd.read_csv("New_York_State_Statewide_COVID-19_Hospitalizations_and_Beds_20250217.csv")
```

```python
c1 = pd.read_csv("Cluster1FinalData.csv")
c2 = pd.read_csv("Cluster2FinalData.csv")
c3 = pd.read_csv("Cluster3FinalData.csv")
```

```python
graphlist = ['Patients Admitted Due to COVID',
       'Patients Admitted Not Due to COVID', 'Patients Newly Admitted',
       'Patients Discharged',
       'Patients Currently in ICU', 'Patients Currently ICU Intubated',
       'Patients Expired','Median Income', 'Less than high school',
       'High school graduate', "Some college or associate's degree",
       "Bachelor's degree or higher",
       'Management, business, science, and arts occupations',
       'Service occupations', 'Sales and office occupations',
       'Natural resources, construction, and maintenance occupations',
       'Production, transportation, and material moving occupations',
       'Unemployed', 'POVERTY']
```

```python
c1namelist = c1["Unnamed: 0"].unique()
c2namelist = c2["Unnamed: 0"].unique()
c3namelist = c3["Unnamed: 0"].unique()
```

```python
datalist = [c1,c2,c3]
featurelist = ['Median Income', 'Less than high school',
       'High school graduate', "Some college or associate's degree",
       "Bachelor's degree or higher",
       'Management, business, science, and arts occupations',
       'Service occupations', 'Sales and office occupations',
       'Natural resources, construction, and maintenance occupations',
       'Production, transportation, and material moving occupations',
       'Unemployed', 'POVERTY']
keylist = ['Wave 3_Patients Admitted Due to COVID',
       'Wave 3_Patients Admitted Not Due to COVID',
       'Wave 3_Patients Newly Admitted', 'Wave 3_Patients Discharged',
       'Wave 3_Patients Currently in ICU',
       'Wave 3_Patients Currently ICU Intubated', 'Wave 3_Patients Expired']
```

```python
#overall regression base on one feature
def overregression(data,y,x):
    x_train,x_test,y_train,y_test = train_test_split(data[x],data[[y]], test_size=0.2,random_state=42)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    y_pred = lr.predict(x_test)
    r2M1 = r2_score(y_test, y_pred)
    mseM1 = mean_squared_error(y_test, y_pred)
    plt.scatter(y_test, y_pred)
    plt.show()
    return r2M1, mseM1
```

```python
for x in datalist:
    print(1)
    for y in keylist:
        r2, mse = overregression(x,y,featurelist)
        print(y)
        print(f"Model 1: R-squared (R²) Score: {r2}")
        print(f"Model1: Mean Squared Error: {mse}")
```

```python
def overregression2(data,y,x):
    x_train,x_test,y_train,y_test = train_test_split(data[x],data[[y]], test_size=0.2,random_state=42)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    y_pred = lr.predict(x_test)
    r2M1 = r2_score(y_test, y_pred)
    mseM1 = mean_squared_error(y_test, y_pred)
    f_stat, p_values = f_regression(x_train, y_train)
    plt.scatter(y_test, y_pred)
    plt.show()
    return r2M1, mseM1,f_stat, p_values
```

```python
manualtestfeatures = ['Median Income','Less than high school',"Bachelor's degree or higher",'Service occupations','Sales and office occupations','Natural resources, construction, and maintenance occupations','Production, transportation, and material moving occupations','Unemployed','POVERTY']
key = 'Patients Currently in ICU'
alpha = 0.05
r2, mse,f_stat, p_values = overregression2(wave3dfc3,key,manualtestfeatures)
print(f"Model 1: R-squared (R²) Score: {r2}")
print(f"Model 1: Mean Squared Error: {mse}")
io = 0
for z in p_values:
    if z > alpha:
        print(z)
        print(manualtestfeatures[io])
        io = io + 1
    else:
        io = io + 1
```
